```python
import pandas as pd

data = {
    'essay': [
        'Education is the most powerful weapon to change the world.',
        'Climate change is an urgent problem we must address.',
        'My cat is cute but not relevant to school topics.',
        'The internet has transformed communication in the 21st century.'
    ],
    'domain_score': [9, 8, 3, 7]
}

df = pd.DataFrame(data)
df.to_csv('essays.csv', index=False)
from google.colab import files
files.download('essays.csv')

```

```python
!pip install -q xgboost transformers datasets
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
nltk.download('stopwords')
nltk.download('punkt')
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("essays.csv")
df = df[['essay', 'domain_score']]
df.dropna(inplace=True)
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = ' '.join([word for word in text.split() if word not in stopwords.words('english')])
    return text
df['clean_essay'] = df['essay'].apply(preprocess_text)
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df['clean_essay']).toarray()
y = df['domain_score']
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))
print("Random Forest RMSE:", rf_rmse)
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
xgb_preds = xgb.predict(X_test)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_preds))
print("XGBoost RMSE:", xgb_rmse)
print(f"RF RMSE: {rf_rmse:.3f}")
print(f"XGBoost RMSE: {xgb_rmse:.3f}")
```
